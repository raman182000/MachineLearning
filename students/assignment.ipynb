```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

```python
data=pd.read_csv("StudentsPerformance.csv")
```

```python
sns.palplot(sns.color_palette('winter_r'))
```

```python
plt.figure(figsize=(8,7))
sns.countplot(data['gender'])
plt.title('Gender Diagram',fontdict={'fontsize':14})
plt.xlabel('Gender')
plt.ylabel('Count')
plt.show()
```

```python
labels=data['gender'].value_counts().index.tolist()
values=data['gender'].value_counts().tolist()
plt.figure(figsize=(8,8))
plt.pie(x=values,labels=labels,autopct="%1.2f%%",shadow=True)
plt.show()
```

```python
plt.figure(figsize=(10,7))
sns.histplot(data['math score'],stat='density')
sns.kdeplot(data['math score'],color='blue')
plt.axvline(data['math score'].mean(),color='red', linestyle='--',linewidth='0.8')
min_ylim,max_ylim=plt.ylim()
plt.text(data['math score'].mean()*1.05,max_ylim*0.96,'mean:{:.2f}'.format(data['math score'].mean()))
plt.xlabel("score")
plt.ylabel("density")
plt.show()
```
